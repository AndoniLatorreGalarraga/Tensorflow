## K Means Visualization

Initialize/create the dataset:

```python
k = 3
p_x, p_y, p_c = [], [], []
for i in range(k):
    cen_x, cen_y = random.randint(0,1000), random.randint(0,1000)
    dev_x, dev_y = random.randint(25, 50), random.randint(50, 150) 
    for j in range(random.randint(25, 45)):
        p_x.append(random.normalvariate(cen_x, dev_x))
        p_y.append(random.normalvariate(cen_y, dev_y))
        p_c.append('#95A4AD')
```

Lets look at the data

```python
plt.scatter(p_x, p_y, c=p_c)
```

We have the first frame

```python
frames = []
frame = [p_x, p_y, p_c]
frames.append(frame)
```
