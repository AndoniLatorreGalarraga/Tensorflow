Predicting from files

```python
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
from tensorflow.keras.applications.resnet50 import preprocess_input, decode_predictions
from tensorflow.keras.preprocessing import image
```

```python
img_pug = image.load_img('D:/Fotos/Guardadas/pugNet.png', target_size=(160, 160))
img_cat = image.load_img('D:/Fotos/Guardadas/catNet.png', target_size=(160, 160))
```

```python
pug_array = image.img_to_array(img_pug)
pug_batch = np.expand_dims(pug_array, axis=0)

cat_array = image.img_to_array(img_cat)
cat_batch = np.expand_dims(cat_array, axis=0)
```

```python
pug_preprocessed = preprocess_input(pug_batch)
cat_preprocessed = preprocess_input(cat_batch)
```

```python
model = tf.keras.models.load_model('D:/models/tensorflow/dogs_vs_cats.h5')
```

```python
pug_prediction = model.predict(pug_preprocessed)
cat_prediction = model.predict(cat_preprocessed)
```

```python
def get_label(n):
    cat_dis = abs(n)
    dog_dis = abs(n - 1)
    if cat_dis < dog_dis:
        return 'cat'
    return 'dog'
```

```python
plt.figure()
plt.imshow(img_pug)
plt.title('Guess: ' + get_label(pug_prediction[0][0]))
```

```python
plt.figure()
plt.imshow(img_cat)
plt.title('Guess: ' + get_label(cat_prediction[0][0]))
```

```python
img = image.load_img('D:/Fotos/Guardadas/cara andoni.png', target_size=(160, 160))
array = image.img_to_array(img)
batch = np.expand_dims(array, axis=0)
preprocessed = preprocess_input(batch)
prediction = model.predict(preprocessed)
plt.figure()
plt.imshow(img)
plt.title('Guess: ' + get_label(prediction[0][0]))
```
